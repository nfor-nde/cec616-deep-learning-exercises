## 1. What Is a Module?

A **module** is any `.py` file containing Python code — functions, classes, variables.  
A **package** is a directory of modules with an `__init__.py` file.

Benefits:
- **Reusability** — write once, import anywhere
- **Namespace separation** — avoid name collisions
- **Organisation** — group related code together

```
my_project/
│
├── main.py              ← uses the modules below
├── math_utils.py        ← custom module
└── utils/
    ├── __init__.py      ← marks utils/ as a package
    ├── string_tools.py
    └── date_tools.py
```

## 2. Import Styles

| Style | Syntax | Access |
|---|---|---|
| Full import | `import math` | `math.sqrt(4)` |
| Named import | `from math import sqrt` | `sqrt(4)` |
| Multiple | `from math import sqrt, pi` | `sqrt(4)`, `pi` |
| Alias | `import numpy as np` | `np.array(...)` |
| Wildcard | `from math import *` | `sqrt(4)` — **avoid in production** |

````python
# Full import
import math
print(math.sqrt(144))   # 12.0
print(math.pi)          # 3.14159...

# Named import
from math import factorial, log2
print(factorial(10))    # 3628800
print(log2(1024))       # 10.0

# Alias
import random as rnd
print(rnd.randint(1, 100))
````

## 3. Standard Library Modules

### 3.1 `math` — Mathematical Functions

````python
import math

print(f"pi          = {math.pi:.6f}")
print(f"e           = {math.e:.6f}")
print(f"sqrt(2)     = {math.sqrt(2):.6f}")
print(f"ceil(3.2)   = {math.ceil(3.2)}")
print(f"floor(3.9)  = {math.floor(3.9)}")
print(f"log10(1000) = {math.log10(1000)}")
print(f"sin(pi/2)   = {math.sin(math.pi/2)}")
print(f"gcd(48, 18) = {math.gcd(48, 18)}")
print(f"comb(10,3)  = {math.comb(10, 3)}")  # 10 choose 3
````

### 3.2 `random` — Randomness

````python
import random

random.seed(42)   # for reproducible results

print(random.randint(1, 10))           # random int in [1, 10]
print(random.random())                  # float in [0.0, 1.0)
print(random.uniform(1.5, 5.5))        # float in [1.5, 5.5]

choices = ["Red", "Green", "Blue", "Yellow"]
print(random.choice(choices))          # one random item
print(random.choices(choices, k=3))    # 3 items with replacement
print(random.sample(choices, k=3))     # 3 items WITHOUT replacement

deck = list(range(1, 14))
random.shuffle(deck)
print("Shuffled deck:", deck)
````

### 3.3 `datetime` — Dates and Times

````python
from datetime import datetime, date, timedelta

now   = datetime.now()
today = date.today()

print(f"Now:   {now}")
print(f"Today: {today}")
print(f"Year:  {now.year}")
print(f"Day of week: {now.strftime('%A')}")
print(f"Formatted: {now.strftime('%d/%m/%Y %H:%M')}")

# Date arithmetic
exam_date    = date(2025, 5, 15)
days_left    = (exam_date - today).days
print(f"\nDays until exam: {days_left}")

# Add days
deadline = today + timedelta(days=14)
print(f"Assignment deadline: {deadline}")
````

### 3.4 `os` and `sys` — System Interaction

## 4. Creating Custom Modules

Any `.py` file is a module. Save the cell below as `math_utils.py` in the same directory as your notebook, then import it.

### Step 1 — Create `math_utils.py`

````python
# Writing math_utils.py to disk so we can import it
module_code = '''
"""math_utils — reusable mathematical helpers."""

PI = 3.141592653589793

def circle_area(r):
    """Return area of a circle with radius r."""
    return PI * r ** 2

def circle_circumference(r):
    return 2 * PI * r

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def clamp(value, lo, hi):
    """Clamp value to the range [lo, hi]."""
    return max(lo, min(value, hi))
'''

with open("math_utils.py", "w") as f:
    f.write(module_code)

print("math_utils.py created.")
````

### Step 2 — Import and Use the Module

````python
import math_utils as mu

r = 5
print(f"Area:          {mu.circle_area(r):.2f}")
print(f"Circumference: {mu.circle_circumference(r):.2f}")
print(f"is_prime(17):  {mu.is_prime(17)}")
print(f"factorial(8):  {mu.factorial(8)}")
print(f"clamp(150, 0, 100): {mu.clamp(150, 0, 100)}")

# Inspect what a module exposes
print("\nPublic names:", [n for n in dir(mu) if not n.startswith("_")])
````

## 5. The `__name__` Guard

When Python runs a file directly, `__name__` equals `"__main__"`.  
When it is imported, `__name__` equals the module's filename (without `.py`).

```python
if __name__ == "__main__":
    # Only runs when script is executed directly
    # NOT when the module is imported
    run_tests()
```

This pattern lets you write modules that also work as standalone scripts.

````python
# In a notebook __name__ is '__main__'
print(f"__name__ = '{__name__}'")

if __name__ == "__main__":
    print("Running as a script (or notebook)")
else:
    print("Running as an imported module")
````

## 6. Packages

A package is a **directory containing an `__init__.py`** file.

```
utils/
├── __init__.py          ← makes utils a package
├── string_tools.py
└── number_tools.py
```

```python
from utils import string_tools
from utils.number_tools import is_prime
```

````python
import os

# Create package structure
os.makedirs("utils", exist_ok=True)

with open("utils/__init__.py", "w") as f:
    f.write('# utils package\n')

with open("utils/string_tools.py", "w") as f:
    f.write('''
def reverse(s):
    return s[::-1]

def is_palindrome(s):
    clean = s.lower().replace(" ", "")
    return clean == clean[::-1]

def word_count(s):
    return len(s.split())
''')

print("Package created.")
````

````python
from utils.string_tools import reverse, is_palindrome, word_count

print(reverse("Python"))                      # nohtyP
print(is_palindrome("A man a plan a canal Panama"))  # True
print(word_count("Hello World from Python"))  # 4
````
