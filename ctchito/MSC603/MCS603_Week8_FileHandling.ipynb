## 1. Opening and Closing Files

```python
f = open("filename.txt", mode)
# ... work with f ...
f.close()   # ALWAYS close — releases the OS handle
```

### File Modes
| Mode | Meaning |
|---|---|
| `"r"` | Read (default) — error if file missing |
| `"w"` | Write — **creates or truncates** |
| `"a"` | Append — creates if missing, adds to end |
| `"x"` | Exclusive create — error if file exists |
| `"r+"` | Read + write |
| `"b"` | Binary mode (combine: `"rb"`, `"wb"`) |

> **Best practice:** always use `with open(...)` — it closes the file automatically, even if an exception occurs.

## 2. Writing Text Files

````python
# Write a simple text file
with open("notes.txt", "w") as f:
    f.write("MCS603 Programming II\n")
    f.write("Week 8: File Handling\n")
    f.write("File I/O is a fundamental skill.\n")

print("notes.txt written.")
````

````python
# Append to an existing file
with open("notes.txt", "a") as f:
    f.write("--- Additional Notes ---\n")
    f.write("Always close your files!\n")

print("Appended to notes.txt.")
````

## 3. Reading Text Files

| Method | Returns |
|---|---|
| `f.read()` | Entire file as one string |
| `f.readline()` | Next line (including `\n`) |
| `f.readlines()` | List of all lines |
| `for line in f` | Iterate line by line (memory-efficient) |

````python
# Read entire file
with open("notes.txt", "r") as f:
    content = f.read()
print(content)
````

````python
# Read as list of lines
with open("notes.txt", "r") as f:
    lines = f.readlines()

print(f"Total lines: {len(lines)}")
for i, line in enumerate(lines, 1):
    print(f"{i:2}: {line.rstrip()}")   # rstrip removes trailing \n
````

````python
# Iterate line by line — best for large files
word_count = 0
with open("notes.txt", "r") as f:
    for line in f:
        word_count += len(line.split())

print(f"Total words: {word_count}")
````

## 4. Working with Paths (`os.path`)

Always build paths portably — never hardcode `/` or `\\`.

````python
import os

# Build paths
data_dir  = os.path.join("data", "students")
file_path = os.path.join(data_dir, "grades.csv")
print("Path:", file_path)

# Inspect paths
print("Exists:",      os.path.exists("notes.txt"))
print("Is file:",     os.path.isfile("notes.txt"))
print("Basename:",    os.path.basename("data/students/grades.csv"))
print("Dirname:",     os.path.dirname("data/students/grades.csv"))
print("Split ext:",   os.path.splitext("grades.csv"))

# Create directory
os.makedirs("output", exist_ok=True)
print("output/ directory ready.")
````

## 5. CSV Files

CSV (Comma-Separated Values) is the most common format for tabular data exchange.

Python's built-in `csv` module handles quoting, escaping, and different delimiters automatically — **always prefer it over manual `split(",")`**.

````python
import csv

students = [
    ["Name",    "ID",     "Math", "English", "Science"],
    ["Alice",   2024001,   92,     88,         95],
    ["Bob",     2024002,   75,     82,         70],
    ["Charlie", 2024003,   88,     90,         85],
    ["Diana",   2024004,   65,     72,         78],
    ["Eve",     2024005,   95,     91,         98],
]

with open("grades.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(students)

print("grades.csv written.")
````

````python
# Read CSV
with open("grades.csv", "r") as f:
    reader = csv.reader(f)
    header = next(reader)   # first row = header
    print("Columns:", header)
    for row in reader:
        print(row)
````

````python
# DictWriter / DictReader — more readable
fieldnames = ["name", "id", "math", "english", "science"]
records    = [
    {"name": "Frank",  "id": 2024006, "math": 80, "english": 75, "science": 88},
    {"name": "Grace",  "id": 2024007, "math": 91, "english": 95, "science": 93},
]

with open("grades2.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(records)

# Read back with DictReader
with open("grades2.csv", "r") as f:
    for row in csv.DictReader(f):
        avg = (int(row["math"]) + int(row["english"]) + int(row["science"])) / 3
        print(f"{row['name']:<10} avg = {avg:.1f}")
````

## 6. Practical Patterns

### Log Writer

````python
from datetime import datetime

def log(message, filepath="app.log"):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(filepath, "a") as f:
        f.write(f"[{timestamp}] {message}\n")

log("Application started")
log("User Alice logged in")
log("Grade record updated: ID=2024001")
log("Application stopped")

with open("app.log") as f:
    print(f.read())
````
