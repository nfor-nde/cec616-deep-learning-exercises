## 1. Python Overview & the Interpreter

Python is a **high-level, interpreted, dynamically typed** language designed for readability.

| Feature | Description |
|---|---|
| Interpreted | Code is executed line-by-line — no compile step needed |
| Dynamically typed | Variable types are resolved at runtime |
| Indentation-based | Block structure is defined by whitespace, not `{}` |
| Batteries included | Rich standard library out of the box |

### Running Python

- **Script mode:** `python my_script.py`

## 2. Syntax Rules: Indentation & Comments

### Indentation
Python uses **consistent indentation** (4 spaces recommended) to define code blocks.  
Mixing tabs and spaces causes a `TabError`.

### Comments
| Type | Syntax | Use |
|---|---|---|
| Single-line | `# comment` | Inline explanation |
| Multi-line | `""" ... """` | Docstrings / block notes |

````python
# Single-line comment

"""
This is a multi-line comment (docstring).
Useful for documenting functions and modules.
"""

# Indentation defines the block — this is a preview of if-statements
x = 10
if x > 5:
    print("x is greater than 5")  # 4-space indent — part of the if block
print("This always runs")          # back at column 0 — outside the block
````

## 3. Variables & Naming Conventions

A **variable** is a named container for a value. Python creates it on first assignment.

### Rules
- Must start with a letter or underscore (`_`)
- Can contain letters, digits, and underscores
- **Case-sensitive:** `age` ≠ `Age` ≠ `AGE`
- Cannot use Python keywords (`if`, `for`, `class`, …)

### Conventions (PEP 8)
| Style | Used for | Example |
|---|---|---|
| `snake_case` | variables & functions | `student_name` |
| `UPPER_CASE` | constants | `MAX_SCORE = 100` |
| `PascalCase` | classes | `StudentRecord` |

````python
# Variable assignment
student_name = "Alice"
student_id   = 2024001
gpa          = 3.75
is_enrolled  = True

# Multiple assignment
x = y = z = 0

# Tuple unpacking
first, second, third = 10, 20, 30

print(student_name, student_id, gpa, is_enrolled)
print(x, y, z)
print(first, second, third)
````

## 4. Data Types

Python has several **built-in data types**:

| Category | Type | Example |
|---|---|---|
| Numeric | `int` | `42` |
| Numeric | `float` | `3.14` |
| Numeric | `complex` | `2 + 3j` |
| Text | `str` | `"hello"` |
| Boolean | `bool` | `True`, `False` |
| None | `NoneType` | `None` |

Use `type()` to inspect the type of any value.

````python
# Checking types
age     = 21
height  = 1.75
name    = "Bob"
passed  = True
nothing = None

print(type(age))      # <class 'int'>
print(type(height))   # <class 'float'>
print(type(name))     # <class 'str'>
print(type(passed))   # <class 'bool'>
print(type(nothing))  # <class 'NoneType'>
````

````python
# String properties
course = "Programming II"

print(len(course))          # length
print(course.upper())       # uppercase
print(course.lower())       # lowercase
print(course[0])            # first character — indexing starts at 0
print(course[-1])           # last character
print(course[0:11])         # slice: characters 0 to 10
print(course.replace("II", "2"))  # replace substring
````

## 5. Type Conversion

Convert between types using built-in functions:

| Function | Converts to |
|---|---|
| `int(x)` | integer |
| `float(x)` | float |
| `str(x)` | string |
| `bool(x)` | boolean |

````python
# Implicit vs explicit conversion
a = 5
b = 2.0

# Implicit: Python widens int → float automatically
result = a + b
print(result, type(result))   # 7.0 <class 'float'>

# Explicit (casting)
print(int(3.9))     # 3  — truncates, does NOT round
print(float("3.14")) # 3.14
print(str(100))     # '100'
print(bool(0))      # False
print(bool(42))     # True
print(bool(""))     # False — empty string is falsy
print(bool("hi"))   # True
````

## 6. Operators

### 6.1 Arithmetic Operators
| Operator | Operation | Example | Result |
|---|---|---|---|
| `+` | Addition | `7 + 3` | `10` |
| `-` | Subtraction | `7 - 3` | `4` |
| `*` | Multiplication | `7 * 3` | `21` |
| `/` | True division | `7 / 3` | `2.333...` |
| `//` | Floor division | `7 // 3` | `2` |
| `%` | Modulus (remainder) | `7 % 3` | `1` |
| `**` | Exponentiation | `2 ** 8` | `256` |

````python
a, b = 17, 5

print("Addition:       ", a + b)   # 22
print("Subtraction:    ", a - b)   # 12
print("Multiplication: ", a * b)   # 85
print("True division:  ", a / b)   # 3.4
print("Floor division: ", a // b)  # 3
print("Modulus:        ", a % b)   # 2
print("Exponentiation: ", 2 ** 10) # 1024

# Practical: check even/odd with modulus
n = 42
print("42 is even:", n % 2 == 0)
````

### 6.2 Comparison Operators
Return `True` or `False`.

| Operator | Meaning |
|---|---|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal |
| `<=` | Less than or equal |

````python
x, y = 10, 20

print(x == y)   # False
print(x != y)   # True
print(x < y)    # True
print(x >= 10)  # True
````

### 6.3 Logical Operators
| Operator | Meaning | Example |
|---|---|---|
| `and` | Both must be True | `x > 0 and x < 100` |
| `or` | At least one True | `x < 0 or x > 100` |
| `not` | Inverts truth value | `not True` → `False` |

````python
age   = 20
score = 75

# Can a student sit an exam? (age >= 18 AND score >= 50)
eligible = age >= 18 and score >= 50
print("Eligible:", eligible)  # True

# Needs remedial support? (score < 50 OR attendance < 75)
attendance = 60
needs_support = score < 50 or attendance < 75
print("Needs support:", needs_support)  # True

# Logical NOT
print(not eligible)  # False
````

### 6.4 Assignment Operators
| Operator | Equivalent to |
|---|---|
| `x += 5` | `x = x + 5` |
| `x -= 5` | `x = x - 5` |
| `x *= 2` | `x = x * 2` |
| `x /= 2` | `x = x / 2` |
| `x //= 2` | `x = x // 2` |
| `x **= 3` | `x = x ** 3` |

````python
balance = 1000
balance += 500   # deposit
print("After deposit:", balance)   # 1500

balance -= 200   # withdrawal
print("After withdrawal:", balance) # 1300

balance *= 1.05  # 5% interest
print("After interest:", round(balance, 2))  # 1365.0
````

## 7. Input / Output

- `print()` — outputs to console; accepts multiple values separated by commas
- `input()` — reads a line from the user and **always returns a string**
- **f-strings** (Python 3.6+) — the preferred way to format output

````python
# print() options
name  = "Alice"
score = 95.5

# Basic
print(name, score)

# Custom separator
print(name, score, sep=" | ")

# f-string (recommended)
print(f"Student: {name}, Score: {score:.1f}")

# format() method
print("Student: {}, Score: {:.1f}".format(name, score))
````
