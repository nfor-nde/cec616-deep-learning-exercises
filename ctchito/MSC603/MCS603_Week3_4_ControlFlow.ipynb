## 1. Conditional Statements

Conditionals let a program make decisions at runtime.

```
if <condition>:
    # executes when condition is True
elif <another_condition>:
    # executes when the first is False but this is True
else:
    # executes when all conditions above are False
```

> **Rule:** Only the first matching branch runs — the rest are skipped.

````python
# Example: Grade classifier
score = 72

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
````

````python
# Combining conditions with logical operators
age    = 20
has_id = True

if age >= 18 and has_id:
    print("Access granted")
elif age >= 18 and not has_id:
    print("Please show your ID")
else:
    print("Access denied — underage")
````

## 2. Nested Conditionals & Ternary Expression

### Nested Conditionals
An `if` block can contain another `if` block inside it. Use sparingly — deep nesting is hard to read.

### Ternary (Conditional) Expression
```python
value_if_true  if  condition  else  value_if_false
```
A one-liner replacement for simple `if/else`.

````python
# Nested: classify a number as positive/negative/zero AND even/odd
n = -8

if n > 0:
    if n % 2 == 0:
        print("Positive even")
    else:
        print("Positive odd")
elif n < 0:
    print("Negative number")
else:
    print("Zero")
````

````python
# Ternary expression
score = 55
result = "Pass" if score >= 50 else "Fail"
print(result)

# Finding the larger of two numbers
a, b = 17, 25
larger = a if a > b else b
print(f"Larger: {larger}")
````

## 3. `for` Loops

A `for` loop **iterates over a sequence** (list, string, range, tuple, …) one item at a time.

```python
for variable in sequence:
    # body executed for each item
```

````python
# Iterating over a string
for char in "Python":
    print(char, end=" ")  # end=" " keeps output on one line
print()  # newline
````

````python
# Iterating over a list
courses = ["MCS601", "MCS602", "MCS603", "MCS604"]

for course in courses:
    print(f"  • {course}")
````

````python
# enumerate() gives both index and value
students = ["Alice", "Bob", "Charlie", "Diana"]

for i, name in enumerate(students, start=1):
    print(f"{i}. {name}")
````

## 4. `range()` in Depth

`range(start, stop, step)` generates a sequence of integers.

| Call | Produces |
|---|---|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(10, 0, -1)` | 10, 9, 8, …, 1 |

> **Note:** `stop` is **exclusive** — `range(1, 6)` does NOT include 6.

````python
# Counting up
print("Counting up:  ", list(range(1, 6)))

# Counting down
print("Counting down:", list(range(5, 0, -1)))

# Even numbers 0–20
print("Even 0-20:    ", list(range(0, 21, 2)))

# Sum of first 100 natural numbers
total = 0
for n in range(1, 101):
    total += n
print(f"Sum 1..100 = {total}")  # 5050 — can verify: n(n+1)/2 = 100*101/2
````

## 5. `while` Loops

A `while` loop runs **as long as a condition is True**. Use it when the number of iterations is not known in advance.

```python
while condition:
    # body
    # must move toward making condition False to avoid infinite loop
```

> **Warning:** Always ensure the loop variable changes so the condition eventually becomes False.

````python
# Countdown timer
count = 5
while count > 0:
    print(count)
    count -= 1   # without this line, the loop would run forever
print("Lift off!")
````

````python
# Accumulate until threshold
total   = 0
counter = 1

while total < 50:
    total += counter
    counter += 1

print(f"Sum reached {total} after adding {counter - 1} numbers")
````

````python
# Input validation pattern (simulated)
valid_inputs = ["yes", "no"]
responses    = ["maybe", "sure", "yes"]  # simulates user typing

response = responses.pop(0)
while response not in valid_inputs:
    print(f"'{response}' is invalid. Please enter 'yes' or 'no'.")
    response = responses.pop(0)

print(f"Valid response received: '{response}'")
````

## 6. Loop Control: `break`, `continue`, `pass`

| Statement | Effect |
|---|---|
| `break` | Exit the loop immediately |
| `continue` | Skip the rest of this iteration; start the next |
| `pass` | Do nothing — placeholder (syntax requires a body) |

````python
# break — stop searching once found
target = 7
numbers = [3, 1, 9, 7, 4, 6, 2]

for i, num in enumerate(numbers):
    if num == target:
        print(f"Found {target} at index {i}")
        break
else:
    # The for-else clause: runs if loop completed WITHOUT break
    print(f"{target} not found")
````

````python
# continue — skip even numbers, print only odd
for n in range(1, 11):
    if n % 2 == 0:
        continue       # skip even
    print(n, end=" ")
print()
````

````python
# pass — placeholder for a block not yet implemented
for n in range(5):
    if n == 3:
        pass   # TODO: handle special case for 3
    print(n, end=" ")
print()
````

## 7. Nested Loops

A loop inside another loop. The **inner loop completes fully** for each iteration of the outer loop.

Total iterations = outer_count × inner_count

````python
# Multiplication table (partial)
print("Multiplication Table (1–5)\n")
print("    ", end="")
for col in range(1, 6):
    print(f"{col:4}", end="")
print("\n" + "-" * 24)

for row in range(1, 6):
    print(f"{row:2} |", end="")
    for col in range(1, 6):
        print(f"{row * col:4}", end="")
    print()
````

````python
# Coordinate grid
for row in range(3):
    for col in range(3):
        print(f"({row},{col})", end="  ")
    print()
````

### 8.1 Factorial  
`n! = n × (n-1) × (n-2) × … × 1`  
`0! = 1` (by definition)

````python
def factorial(n):
    """Return n! using a for loop."""
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

for n in range(0, 11):
    print(f"{n:2}! = {factorial(n)}")
````

### 8.2 Fibonacci Sequence  
Each number is the sum of the previous two: `0, 1, 1, 2, 3, 5, 8, 13, 21, …`

````python
def fibonacci(n):
    """Return a list of the first n Fibonacci numbers."""
    if n <= 0:
        return []
    if n == 1:
        return [0]

    seq = [0, 1]
    while len(seq) < n:
        seq.append(seq[-1] + seq[-2])
    return seq

print(fibonacci(15))
````

````python
# Check if a number is Fibonacci
def is_fibonacci(num):
    a, b = 0, 1
    while b < num:
        a, b = b, a + b
    return b == num or num == 0

for n in [0, 1, 4, 8, 13, 21, 22]:
    print(f"{n:3} is Fibonacci: {is_fibonacci(n)}")
````

### 8.3 Number Patterns

Print triangle patterns using nested loops.

````python
# Right-angle triangle of stars
n = 5
print("Right-angle triangle:")
for i in range(1, n + 1):
    print("*" * i)
````

````python
# Number triangle
print("Number triangle:")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
````

````python
# Centered pyramid
n = 5
print("Pyramid:")
for i in range(1, n + 1):
    spaces = " " * (n - i)
    stars  = "* " * i
    print(spaces + stars)
````

````python
# Diamond
n = 4
print("Diamond:")
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "* " * i)
````

### 8.4 Prime Numbers

A number is **prime** if it is greater than 1 and has no divisors other than 1 and itself.

````python
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

primes = [n for n in range(2, 51) if is_prime(n)]
print("Primes up to 50:", primes)
````

Build a simple calculator menu that runs until the user chooses "Exit".

Menu options: `1-Add  2-Subtract  3-Multiply  4-Divide  5-Exit`

Use a simulated sequence of inputs to avoid blocking the notebook.
