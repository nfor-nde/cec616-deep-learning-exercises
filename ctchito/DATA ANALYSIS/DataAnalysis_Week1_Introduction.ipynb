## 1. What Is Data Analysis?

**Data analysis** is the process of inspecting, cleaning, transforming, and modelling data to discover useful information, inform conclusions, and support decision-making.

### Why It Matters
| Domain | Example Decision Supported |
|---|---|
| Healthcare | Which treatment leads to better patient outcomes? |
| Business | Which product line is most profitable? |
| Government | Which regions need more infrastructure investment? |
| Sports | Which player recruitment will improve team performance? |
| Education | Which teaching method improves student pass rates? |

### Types of Data Analysis
| Type | Question Answered | Example |
|---|---|---|
| **Descriptive** | What happened? | Sales were up 12% in Q3 |
| **Diagnostic** | Why did it happen? | Q3 growth driven by new market entry |
| **Predictive** | What will happen? | Q4 sales forecast: $2.3M |
| **Prescriptive** | What should we do? | Increase marketing spend in Region B |

## 2. Types of Data

### 2.1 By Structure
```
Data
├── Structured       → rows & columns (tables, spreadsheets, SQL)
├── Semi-structured  → key-value pairs (JSON, XML, logs)
└── Unstructured     → free-form (text, images, audio, video)
```

### 2.2 By Measurement Scale
| Scale | Type | Examples | Operations |
|---|---|---|---|
| **Nominal** | Categorical | Country, Gender, Blood type | = ≠ |
| **Ordinal** | Ranked | Education level, Star rating | = ≠ < > |
| **Interval** | Numeric, no true zero | Temperature (°C), Year | + − |
| **Ratio** | Numeric, true zero | Height, Income, Age | + − × ÷ |

````python
import pandas as pd
import numpy as np

# Demonstrate different data types in a single DataFrame
sample = pd.DataFrame({
    # Nominal
    "country"      : ["Ghana", "Kenya", "Nigeria", "Egypt", "Senegal"],
    "blood_type"   : ["A", "O", "B", "AB", "A"],
    # Ordinal
    "edu_level"    : ["Secondary", "Tertiary", "Primary", "Tertiary", "Secondary"],
    "health_rating": [3, 5, 2, 4, 3],   # 1–5 Likert scale
    # Interval
    "year_enrolled": [2020, 2019, 2021, 2018, 2022],
    # Ratio
    "age"          : [28, 35, 22, 41, 19],
    "income_usd"   : [1200, 3500, 800, 5200, 600],
})

print("Sample dataset:")
print(sample.to_string(index=False))
print("\nData types:")
print(sample.dtypes)
````

## 3. The Data Analysis Workflow

```
┌─────────────────────────────────────────────────────────┐
│  1. DEFINE      State the question clearly              │
│       ↓                                                  │
│  2. COLLECT     Gather raw data from sources            │
│       ↓                                                  │
│  3. CLEAN       Handle missing values, errors, outliers │
│       ↓                                                  │
│  4. EXPLORE     Descriptive stats, initial plots        │
│       ↓                                                  │
│  5. ANALYSE     Statistical tests, models               │
│       ↓                                                  │
│  6. VISUALISE   Charts, dashboards, reports             │
│       ↓                                                  │
│  7. COMMUNICATE Insights → decisions                    │
└─────────────────────────────────────────────────────────┘
```

Each week of this course covers one or more stages of this workflow.

## 5. Your First Analysis — Africa Health Dataset

We'll use a **54-country Africa Health dataset** (2022) throughout this course.  
This first analysis touches every step of the workflow in miniature.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams.update({"figure.dpi": 110, "axes.grid": True, "grid.alpha": 0.3})
np.random.seed(42)

N = 54
countries = [
    "Nigeria","Ethiopia","Egypt","DR Congo","Tanzania","South Africa","Kenya",
    "Uganda","Algeria","Sudan","Morocco","Mozambique","Ghana","Angola","Cameroon",
    "Madagascar","Côte d'Ivoire","Niger","Burkina Faso","Mali","Malawi","Zambia",
    "Senegal","Chad","Somalia","Zimbabwe","Guinea","Rwanda","Benin","Burundi",
    "Tunisia","South Sudan","Togo","Sierra Leone","Libya","Congo","Liberia",
    "Mauritania","Eritrea","Namibia","Gambia","Botswana","Gabon","Lesotho",
    "Guinea-Bissau","Equatorial Guinea","Mauritius","Eswatini","Djibouti",
    "Comoros","Cabo Verde","Sao Tome","Seychelles","São Tomé"
]
regions = (["West Africa"]*16+["East Africa"]*14+
           ["North Africa"]*6+["Central Africa"]*8+["Southern Africa"]*10)

df = pd.DataFrame({
    "country"           : countries,
    "region"            : regions,
    "life_expectancy"   : np.clip(np.random.normal(63,8,N),45,80),
    "infant_mortality"  : np.clip(np.random.exponential(35,N),5,120),
    "maternal_mortality": np.clip(np.random.exponential(350,N),20,2000),
    "hiv_prevalence"    : np.clip(np.random.exponential(3.5,N),0.1,28),
    "health_expenditure": np.clip(np.random.normal(5.2,2.1,N),1,12),
    "gdp_per_capita"    : np.clip(np.exp(np.random.normal(7.2,1.2,N)),300,15000),
    "vaccination_dtp3"  : np.clip(np.random.normal(78,18,N),20,99),
    "water_access"      : np.clip(np.random.normal(68,22,N),15,99),
})
df.to_csv("africa_health.csv", index=False)
print("Dataset created:", df.shape)
df.head()
````

````python
# ── Step 1: Define the question ──────────────────────────────────────
print("RESEARCH QUESTION")
print("  How does life expectancy vary across African regions,")
print("  and what economic/health factors are associated with it?")
print()

# ── Step 2: Summarise the data ───────────────────────────────────────
print("DATASET SUMMARY")
print(f"  Rows: {df.shape[0]}  Columns: {df.shape[1]}")
print(f"  Missing values: {df.isnull().sum().sum()}")
print()

# ── Step 3: Descriptive statistics ──────────────────────────────────
print("LIFE EXPECTANCY STATISTICS")
le = df["life_expectancy"]
print(f"  Mean:    {le.mean():.1f} years")
print(f"  Median:  {le.median():.1f} years")
print(f"  Std Dev: {le.std():.1f} years")
print(f"  Range:   {le.min():.1f} – {le.max():.1f} years")
print()

# ── Step 4: Quick visualisation ──────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

region_means = df.groupby("region")["life_expectancy"].mean().sort_values()
axes[0].barh(region_means.index, region_means.values,
             color="#3498DB", alpha=0.8, edgecolor="white")
axes[0].set_title("Mean Life Expectancy by Region", fontweight="bold")
axes[0].set_xlabel("Years")

axes[1].scatter(df["gdp_per_capita"], df["life_expectancy"],
                alpha=0.7, color="#E74C3C", edgecolors="white")
axes[1].set_title("GDP vs Life Expectancy", fontweight="bold")
axes[1].set_xlabel("GDP per Capita (USD)")
axes[1].set_ylabel("Life Expectancy (years)")

plt.suptitle("Week 1 — First Look at Africa Health Data", fontsize=13)
plt.tight_layout()
plt.savefig("week1_first_analysis.png", bbox_inches="tight")
plt.show()

print("INITIAL INSIGHT")
best  = region_means.idxmax()
worst = region_means.idxmin()
print(f"  {best} has the highest average life expectancy ({region_means[best]:.1f} yrs)")
print(f"  {worst} has the lowest  average life expectancy ({region_means[worst]:.1f} yrs)")
````
